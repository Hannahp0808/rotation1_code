```python
import os
import re
```

```python
root_dir = "../aligned_files"

resolution_pattern = re.compile(
    r"REMARK\s+2\s+RESOLUTION\.\s+([\d\.]+)\s+ANGSTROMS",
    re.IGNORECASE
)
```

```python
best_resolution = float("inf")
best_complex = None
```

```python
for complex_id in os.listdir(root_dir):
    complex_dir = os.path.join(root_dir, complex_id)
    if not os.path.isdir(complex_dir): #skip if the file path doesn't exist
        continue

    pdb_path = os.path.join(complex_dir, f"{complex_id}.pdb")
    if not os.path.isfile(pdb_path):
        continue  # skip if expected complex pdb is missing

    with open(pdb_path, "r") as f:
        for line in f:
            match = resolution_pattern.search(line)
            if match:
                resolution = float(match.group(1))
                if resolution < best_resolution:
                    best_resolution = resolution
                    best_complex = pdb_path
                break
```

```python
print("Best complex:")
print(best_complex)
print("Resolution (Å):", best_resolution)
```
